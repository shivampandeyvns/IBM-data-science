```python
import numpy as np
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import statistics
```

<h2>Getting the Data</h2>

```python
df=pd.read_csv('concrete_data.csv')
```

```python
df.head()
```

<h2>Splitting the Predictors and the target value columns</h2>

```python
target=df['Strength']
predictors=df[['Cement','Blast Furnace Slag','Fly Ash','Water','Superplasticizer','Coarse Aggregate','Fine Aggregate','Age','Strength']]
```

<h2>Normalizing the data</h2>

```python
predictors_norm=(predictors-predictors.mean())/predictors.std() 
```

<h2>Defining the Regression model function</h2>

```python
def regression_model():
    model=Sequential()
    model.add(Dense(10,activation='relu',input_shape=(predictors.shape[1],)))
    model.add(Dense(10,activation='relu'))
    model.add(Dense(10,activation='relu'))
    model.add(Dense(1))
    
    model.compile(optimizer='adam',loss='mean_squared_error')
    
    return model
```

<h2>Calling the Constructor</h2>

```python
model=regression_model()
```

<h2>Fitting the model to the function</h2>

```python
mean_error=[]
for i in range(50):
    X_train,X_test,Y_train,Y_test=train_test_split(predictors_norm,target,test_size=0.3)#doing a train test split on the data
    model=regression_model()# instantiating the regression model
    model.fit(X_train,Y_train,epochs=100,validation_split=0.3)#fitting the data to the model
    y_pred=model.predict(X_test)#getting the predictions on the test data
    mean_error.append(mean_squared_error(Y_test,y_pred))# appending the mean squared error from all the iterations
```

<h2>Mean and std of the Mean squared error</h2>

```python
mean=sum(mean_error)/len(mean_error)
std=statistics.stdev(mean_error)
```

```python
print('The mean of the Mean squared error is: {}'.format(mean))
print('The Standard Deviation of the Mean squared error is: {}'.format(std))
```
